```python
raw_data = open("./data.txt","r") 
splitted_text = raw_data.read().split('\n\n')
raw_data.close()
```

```python
splitted_text
```

```python
mapping = list()
for line in splitted_text:
    mapping.append(line.split('\n'))
```

```python
seed_mapping = set()
for seed in mapping[0][0].split(' ')[1:]:
    seed_mapping.add(int(seed))
```

```python
seed_to_soil_mapping = mapping[1][1:]
soil_to_fertilizer_mapping = mapping[2][1:]
fertilizer_to_water_mapping = mapping[3][1:]
water_to_light_mapping = mapping[4][1:]
light_to_temperature_mapping = mapping[5][1:]
temperature_to_humidity_mapping = mapping[6][1:]
humidity_to_location_mapping = mapping[7][1:]
```

```python
all_mappings = list()
all_mappings.append(seed_to_soil_mapping)
all_mappings.append(soil_to_fertilizer_mapping)
all_mappings.append(fertilizer_to_water_mapping)
all_mappings.append(water_to_light_mapping)
all_mappings.append(light_to_temperature_mapping)
all_mappings.append(temperature_to_humidity_mapping)
all_mappings.append(humidity_to_location_mapping)
```

```python
map_names = [
    'seed_to_soil_mapping',
    'soil_to_fertilizer_mapping',
    'fertilizer_to_water_mapping',
    'water_to_light_mapping',
    'light_to_temperature_mapping',
    'temperature_to_humidity_mapping',
    'humidity_to_location_mapping',
]
```

```python
all_mappings_dict = dict()
i=0
for maps in all_mappings:
    destination = list()
    source = list()
    step = list()

    for interval in maps:
        destination.append(int(interval.split(' ')[0]))
        source.append(int(interval.split(' ')[1]))
        step.append(int(interval.split(' ')[2]))

    maps = [destination, source, step]
    
    all_mappings_dict[map_names[i]] = maps
    i+=1

del i
```
